# 🧠 Single Agent Pipeline Project

## Problem Statement
Build a **Single-Agent Smart Assistant** that:
- Understands user queries
- Routes tasks based on intent
- Uses tools when required
- Returns structured JSON output

### The agent should handle:
- Math queries → Calculator Tool
- Keyword extraction → Keyword Tool
- General queries → Direct response

---
### 🛠️ What You Need to Implement
- Agent logic
- Conditional routing
- Tool integration
- Basic error handling

### 🚀 Bonus
- Improve routing
- Add logging
- Add more tools


````python
# tool 1: calculator

def calculator(expression: str) -> str:
    """evaluate a mathematical expression."""
    try:
        return str(eval(expression))
    except Exception:
        return "Error in calculation"
````

````python
# tool 2: keyword extractor

def extract_keywords(text: str) -> list:
    """extract keywords from text."""
    try:
        words = text.split()
        keywords = list(set([w.lower() for w in words if len(w) > 4]))
        return keywords[:5]
    except Exception:
        return []
````

##  Implemented Agent Logic Below

 Used conditional routing:
- If query contains "calculate" → use calculator
- If query contains "keywords" → use keyword extractor
- Else → general response

````python
# agent function

def agent(query: str):
    #  input validation
    try:
        if query is None or str(query).strip() == "":
            print("[log] received: empty/none input")
            print("[log] route: error")
            return {"type": "error", "result": "Empty or invalid input provided."}

        query = str(query).strip()
        query_lower = query.lower()
        print(f"[log] received query: {query}")

        #   calculator
        if "calculate" in query_lower:
            print("[log] route: calculation")
            print("[log] tool: calculator")
            # remove trigger words to extract the math expression
            expression = query_lower.replace("calculate", "").strip()
            print(f"[log] expression: {expression}")
            result = calculator(expression)
            print(f"[log] status: success")
            return {"type": "calculation", "result": result}

        #   keyword extraction
        elif "keywords" in query_lower:
            print("[log] route: keywords")
            print("[log] tool: extract_keywords")
            # extract actual text after common preamble phrases
            text = query
            for phrase in ["extract keywords from", "keywords from", "keywords of", "keywords in", "keywords","Extract keywords from","extract keyword from"]:
                if phrase in query_lower:
                    idx = query_lower.index(phrase) + len(phrase)
                    text = query[idx:].strip()
                    break
            print(f"[log] text: {text}")
            result = extract_keywords(text)
            print(f"[log] status: success")
            return {"type": "keywords", "result": result}

        #   general response
        else:
            print("[log] route: general")
            print("[log] status: success")
            return {"type": "general", "result": f"I received your query: '{query}'. No specialized tool was needed."}

    except Exception as e:
        print(f"[log] route: error")
        print(f"[log] exception: {e}")
        return {"type": "error", "result": f"An error occurred: {str(e)}"}
````

## 📦 Output Format

```
{
  "type": "calculation / keywords / general / error",
  "result": ...
}
```

````python
# test cases

queries = [
    "Calculate 20 + 5",
    "Extract keywords from Artificial Intelligence is transforming industries",
    "What is machine learning?"
]

for q in queries:
    print("query:", q)
    print("response:", agent(q))
    print("-" * 50)
````

````python
# interactive mode

while True:
    user_input = input("enter query (type 'exit' to stop): ")
    if user_input.lower() == "exit":
        break
    print("response:", agent(user_input))
````
